```python
import seaborn as sns
tips = sns.load_dataset("tips")
tips.head()
```

```python
tips = sns.load_dataset('tips')
sns.jointplot(x = "total_bill", y = 'tip', data = tips)
```

```python
sns.barplot(x='day', y='total_bill', hue='sex', data= tips)
```

```python
titanic = sns.load_dataset("titanic")
sns.countplot('class', hue = 'who', data = titanic)
```

```python
sns.boxplot('day','total_bill',hue = 'sex', data= tips)
```

```python
sns.pointplot('time','total_bill',hue='smoker',  data= tips, dodge=True )
```

```python
sns.violinplot('day','total_bill', hue ='smoker', data= tips, palette='muted')
```

```python
sns.violinplot('day','total_bill', hue= 'sex',
              data = tips, palette= 'Set2', split = True,
              scale = 'count',inner = 'quartile')
```

```python
sns.stripplot('day','total_bill', hue= 'smoker',
             data = tips, jitter = True,
             palette= 'Set2',split=True)
```

```python
sns.boxplot('tip','day',data=tips, whis = np.inf)
```

```python
sns.stripplot('tip','day', data= tips, jitter = True, color = '.3')
```

```python
sns.boxplot('tip','day',data=tips, whis = np.inf)
sns.stripplot('tip','day', data= tips, jitter = True, color = '.3')
```

```python
sns.swarmplot('day','total_bill',hue= 'sex', data= tips)
```

```python
sns.violinplot('day','total_bill',data= tips, inner = None)
sns.swarmplot('day','total_bill', data=tips, color ='white', edgecolor = 'gray')
```
