** How many people have the word Chief in their job title? (This is pretty tricky) **

```python
count = 0
for i in df["JobTitle"].apply(lambda x : x.lower()) :
    if 'chief' in i :
        count += 1
print(count)
```

```python
df["JobTitle"].apply(lambda x : x.lower()).str.contains('chief').sum()
```

** Bonus: Is there a correlation between length of the Job Title string and Salary? **

```python
title_len = []

for i in df["JobTitle"]:
    title_len.append(len(i))
    

df["Titl_len"] = title_len
df.head()
    
```

```python
df.Titl_len.corr(df.TotalPayBenefits)
```

```python
df["Titl_len"].corr(df['TotalPayBenefits'])
```
