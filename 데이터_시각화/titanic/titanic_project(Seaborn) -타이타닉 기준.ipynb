```python
#map   column
titanic_data['alive'] = titanic_data.survived.map({0:"no", 1:"yes"})

titanic_data.head()
```

```python
sns.catplot('alive', data= titanic_data,kind='count')
```

```python
sns.factorplot('pclass', 'survived', data= titanic_data, order=[1,2,3])
```

```python
sns.catplot('pclass','survived', hue='who', data= titanic_data, order =[1,2,3],dodge=True , kind = 'point' )
```

```python
sns.factorplot('pclass','survived', hue='who', data= titanic_data, order =[1,2,3],dodge=True )
```

```python
sns.factorplot('pclass','survived', hue='who', data= titanic_data, order =[1,2,3],dodge=False )
```
